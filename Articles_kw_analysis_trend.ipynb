```python
import pandas as pd
import plotly.graph_objects as go
```

```python
month_dico = {1: 'Jan',
             2: 'Feb',
             3: 'Mar',
             4: 'Apr',
             5: 'May',
             6: 'Jun',
             7: 'Jul',
             8: 'Aug',
             9: 'Sep',
             10: 'Oct',
             11: 'Nov',
             12: 'Dec'}
```

```python
def plot_evolution_score(df, keyword1, keyword2):
    df1 = df[df['unique_kw'].str.contains(keyword1)]
    grouped_df = df1.groupby('month', as_index = False).agg({'debate_score':'mean'})
#    grouped_df = grouped_df.sort_values(by = 'month', inplace = True)
#    print(grouped_df)
    grouped_df['month'] = grouped_df['month'].map(month_dico)
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=grouped_df['month'], y=grouped_df['debate_score']*1000,
                    name=keyword1))
    df2 = df[df['unique_kw'].str.contains(keyword2)]
    grouped_df = df2.groupby('month', as_index = False).agg({'debate_score':'mean'})
    grouped_df['month'] = grouped_df['month'].map(month_dico)
    fig.add_trace(go.Scatter(x=grouped_df['month'], y=grouped_df['debate_score']*1000,
                name=keyword2))
    return fig
```

```python
plot_evolution_score(articles_df, 'trump', 'biden')
```

```python
kw_ds_dict = {}

for kw in lok_df['keyword']:
    print(kw)
    if len(articles_df[articles_df['unique_kw'].str.contains(str(kw))]) != 0:
        kw_df = articles_df[articles_df['unique_kw'].str.contains(str(kw))]
        kw_ds_dict[kw] = kw_df.debate_score.mean()
    else:
        kw_ds_dict[kw] = 0
```

```python
kw_ds_dict
```

```python
dict_occ = {}

for kw in kw_ds_dict.keys():
    dict_occ[kw] = list(lok_df['nb_occ'])[list(lok_df['keyword']).index(kw)]
```

```python
import numpy as np

d = {'keyword': list(kw_ds_dict.keys()), 'debate_score': list(kw_ds_dict.values()), 'nb_occ': list(dict_occ.values())}

sumup_df = pd.DataFrame(data = d)
```

```python
sumup_df.sort_values(by = 'debate_score', ascending = False).head(50)
```

```python
import plotly.express as px
fig = px.scatter(sumup_df, x='debate_score', y='nb_occ', hover_name = 'keyword')
fig.show()
```
